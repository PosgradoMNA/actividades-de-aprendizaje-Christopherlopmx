How to deal with missing data?

Drop data

Drop the whole column

Replace data

Replace it by mean

Replace it by frequency

Replace it based on other functions

Whole columns should be dropped only if most entries in the column are empty. In our dataset, none of the columns are empty enough to drop entirely. We have some freedom in choosing which method to replace data; however, some methods may seem more reasonable than others. We will apply each method to many different columns:


Replace by mean:


"normalized-losses": 41 missing data, replace them with mean

"stroke": 4 missing data, replace them with mean

"bore": 4 missing data, replace them with mean

"horsepower": 2 missing data, replace them with mean

"peak-rpm": 2 missing data, replace them with mean

Replace by frequency:


"num-of-doors": 2 missing data, replace them with "four".
Reason: 84% sedans is four doors. Since four doors is most frequent, it is most likely to occur
Drop the whole row:

"price": 4 missing data, simply delete the whole row
Reason: price is what we want to predict. Any data entry without price data cannot be used for prediction; therefore any row now without price data is not useful to us

DATA NORMALIZATION:

```python
"""
Target: would like to normalize those variables so their value ranges from 0 to 1

Approach: replace original value by (original value)/(maximum value)
"""

# replace (original value) by (original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```

```python
# Aquí vamos a normalizar la columna height:
df['height'] = df['height']/df['height'].max() # dividimos entre el valor máximo todo, entonces los normalizamos, es esos de los saldos que puse arriba

# show the scaled columns
df[["length","width","height"]].head()
```

Binning

```python
"""
Why binning?
Binning is a process of transforming continuous numerical variables into discrete categorical 'bins' for grouped analysis. (bins son como botes, como que todo esta dividido por categorías/botes/bins como altos, medios, bajos)

Example:

In our dataset, "horsepower" is a real valued variable ranging from 48 to 288 and it has 59 unique values. 
What if we only care about the price difference between cars with high horsepower, medium horsepower, and little horsepower (3 types)?  <--- Esto es lo importante! Esos son 3 bins
Can we rearrange them into three ‘bins' to simplify analysis?

We will use the pandas method 'cut' to segment the 'horsepower' column into 3 bins.
"""
```

```python
df["horsepower"]=df["horsepower"].astype(int, copy=True)  # primero lo vamos a hacer int
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot


# draw historgram of attribute "horsepower" with bins = 3 <--- lo más cool de todo esto pienso yo es que no tienes que ni pensarle, ellos te lo separan
plt.pyplot.hist(df["horsepower"], bins = 3) #<-- aquí le pusimos 3, si le pones otro numero te dice "no hya problema, hacemos N bins"

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

Indicator Variable (or Dummy Variable)

```python
"""
What is an indicator variable?
An indicator variable (or dummy variable) is a numerical variable used to label categories. They are called 'dummies' because the numbers themselves don't have inherent meaning.

Why we use indicator variables?

We use indicator variables so we can use categorical variables for regression analysis in the later modules.

Example
We see the column "fuel-type" has two unique values: "gas" or "diesel". Regression doesn't understand words, only numbers. To use this attribute in regression analysis, we convert "fuel-type" to indicator variables.

We will use pandas' method 'get_dummies' to assign numerical values to different categories of fuel type.
"""

# estas son muy fáciles de entender, son las que te pasas usando tu como flags, para categorizaro como 1 o 0 una variable y con eso poder hacer querys después.
```

```python
df.columns
```

```python
dummy_variable_1 = pd.get_dummies(df["fuel-type"]) #<-- esta cosa las asigna igual, no tienes que ni poner if match=diesel,1,0 o algo así
dummy_variable_1.head()
```

```python
# arriba esta muy simple, si es gas es 1, si es diesel es 0
# ahora vamos a cambiarle los nombres a las columnas y recuerda que para que se cambien debe decir Inplace=True si no no pasa nada
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
# merge data frame "df" and "dummy_variable_1"   # esto de aquí es para juntar las tablas, para que aparezcan los 0 y 1 en columnas dentro de la tabla
df = pd.concat([df, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True)
```
