Calcula la estadística descriptiva con describe() y explica las medidas de tendencia central y dispersión


```python
dataset.describe()
```

```python
import seaborn as sns
```

```python
#Para describir los valores primero voy hacer una gráfica de dispersión de los valores descritos:
describe_num_df = dataset.describe()
num_col = dataset._get_numeric_data().columns

describe_num_df.reset_index(inplace=True)
# To remove any variable from plot
#describe_num_df = describe_num_df[describe_num_df['index']!= 'count']]
for i in num_col:
  if i in ['index']:
    continue
  sns.catplot(x='index', y=i, data=describe_num_df)
  plt.show()
```

```python
#Ahora vamos a hacer una lista de columnas sin considerar ID y Y para trabajar los valores de tendencia central y dispersion
new_num_col=num_col.drop('ID','Y')
```

```python
#Voy a usar X1 como test y para comprar luego con un loop para todos los valores en el index que cree arriba
dataset['X1'].mean()
```

```python
dataset['X1'].median()
```

```python
dataset['X1'].mode()
```

```python
#Este loop es para calcular el primer valor de tendencia central de dispersion, el promedio
for i in new_num_col:
  m = dataset[i].mean()
  print("Para la columna:",i)
  print("Su valor de promedio es:",m)
```

```python
#Ahora la mediana

for p in new_num_col:
  me = dataset[p].median()
  print("Para la columna:",p)
  print("Su valor de mediana es:",me)
```

```python
#Ahora la moda
for l in new_num_col:
  mo = dataset[l].mode()
  print("Para la columna:",l)
  print("Su valor de moda es:",mo)
```

```python
# De forma interesante, en la última medida, las modas son 0, eso quiere decir que la observación que más se repite en todas esas es 0

#Peero luego de leer la documentacion de los datos, esto es porque primero debí haber hecho un mapeo de datos a lo que nos dicen que cada columna representa
#ahora mismo los datos así como estan no representan nada para nosotros, se han convertido todas las variables categoricas a variables numéricas, por ejemplo
#del tipo booleano, un ejemplo es la X2 que describe género:
```
